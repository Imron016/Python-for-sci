## 3. ตัวแปรแบบ global and local

```python
x=1 
def func( ):
    x=10
    return None
func()
print(x) # x=1 and x=10 are different level
```

### local function 

```python
def func():
    global x
    x=10
    return None
func()
print(x)
```

```python
def func ():
    def func1():
        print('in func1')
        return None
    return None
func()
func1()  #it's error cause it different levels
```

```python
def func ():
    global func1 #chang it to global function
    def func1():
        print('in func1')
        return None
    return None
func()
func1()
```

```python
def func(x): 
    x = 10 
x = 1 #this x=1 is highter than that x=10
func(x) 
print(x) 
```

## 5. generator

```python
def cube_list(n):
    for i in range(n):
        yield i**3  #case of make generator list we could use 'yield'
gen=cube_list(5) #yield คงสถานะ เก่าของ generator ไว้
```

```python
next(gen)
```

```python
next(gen),next(gen),next(gen),next(gen)
```

```python
gen1=(x**3 for x in range(10) )
type(gen1)
```
