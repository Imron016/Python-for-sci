```python
x = np.linspace(0,1,10)
y1 = [i**2 for i in x]
y2 = [0.75 * i**2 for i in x]
y3 = [1.25 * i**2 for i in x]
```

```python
plt.plot(x, y1, 'r')
plt.plot(x, y2, c='#4433ff')
plt.plot(x, y3, color=(0.1, 0.2, 0.4))
```

```python
plt.plot(x, y1, linestyle='solid', linewidth=3)
plt.plot(x, y2, ls='-.', lw=6)
plt.plot(x, y3, 'g--')
```

```python
plt.plot(x, y1, linestyle='None', marker='o', markersize=4)
plt.plot(x, y2, ls='None', marker='^', ms=2)
plt.plot(x, y3, ls='None', marker='d', ms=6)
```

```python
x = np.linspace(0,1,10)
y1 = [i**2 for i in x]
y2 = [0.75 * i**2 for i in x]
y3 = [1.25 * i**2 for i in x]
```

```python
plt.plot(x, y1, 'r',label='x**2')
plt.plot(x, y2, c='#4433ff',label='0.75*x**2')
plt.plot(x, y3, color=(0.1, 0.2, 0.4), label='1.25*x**2')
plt.title('Title') #ชื่อกราฟ
plt.xlabel('x') #x axis
plt.ylabel('f(x)')#y axis
plt.xlim([0,1.5]) #boundary x
plt.ylim([0,1.5]) #boundary y
plt.legend(loc=2) #description
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.linspace(0,10,100)
```

```python
y0=[i**2 for i in x]
y1=[0.75*i**2 for i in x]
y2=[1.25*i**2 for i in x]
```

```python

fig, (ax0, ax1, ax2) = plt.subplots(ncols=3, figsize=(12, 4))
fig.set_facecolor('#ffffff') #สีพื้นหลัง
ax0.plot(x,y0)
ax1.plot(x,y1)
ax2.plot(x,y2)
```

```python
fig, (ax0, ax1, ax2) = plt.subplots(nrows=3, figsize=(4,12))
fig.set_facecolor('#0000ff') #สีพื้นหลัง
ax0.plot(x,y0)
ax1.plot(x,y1)
ax2.plot(x,y2)
```
